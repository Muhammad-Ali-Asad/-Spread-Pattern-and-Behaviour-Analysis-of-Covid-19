```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file="COVID-19 Daily.xlsx"
#For Pakistan
data=pd.read_excel(file,sheet_name=["PK"])
data

# Access a specific sheet
data = data['PK']
```

```python
# question1
null_values = data.isnull().sum()
column_info = data.info()
descriptive_stats = data.describe()

null_values, column_info, descriptive_stats

```

```python
# Fill missing values using forward fill and backward fill
data.fillna(method='ffill', inplace=True)
data.fillna(method='bfill', inplace=True)

# Re-check for missing values
print(data.isnull().sum())
```

```python
# question02

total_tests = data['Daily Tests'].sum()
print(f'Total Tests Performed till September 01, 2020: {total_tests}')
```

```python

# Trend Line Graph
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Daily Tests'], label='Daily Tests', color='blue')
plt.title('Daily Testing Trend in Pakistan')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()

# Scatter Diagram
plt.figure(figsize=(12, 6))
plt.scatter(data['Date'], data['Daily Tests'], color='purple')
plt.title('Scatter Plot of Daily Tests')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Bar Graph
plt.figure(figsize=(12, 6))
plt.bar(data['Date'], data['Daily Tests'], color='orange')
plt.title('Bar Graph of Daily Tests')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


average_tests = data['Daily Tests'].mean()

# Above-average in red, below-average in green
colors = ['red' if x > average_tests else 'green' for x in data['Daily Tests']]

plt.figure(figsize=(12, 6))
plt.scatter(data['Date'], data['Daily Tests'], color=colors)
plt.title('Above-Average (Red) and Below-Average (Green) Daily Testing')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# question03

# Convert the dataset into weekly stats
weekly_stats = data.resample('W-Mon', on='Date').sum().reset_index().sort_values(by='Date')

# Convert the dataset into monthly stats
monthly_stats = data.resample('M', on='Date').sum().reset_index().sort_values(by='Date')

# Save the weekly and monthly stats
weekly_stats.to_excel('Weekly_StatsPAK.xlsx', index=False)
monthly_stats.to_excel('Monthly_StatsPAK.xlsx', index=False)

# Display max and min values for weekly and monthly stats
print("Weekly Stats - Max:")
print(weekly_stats.max())

print("Weekly Stats - Min:")
print(weekly_stats.min())

print("Monthly Stats - Max:")
print(monthly_stats.max())

print("Monthly Stats - Min:")
print(monthly_stats.min())
```

```python
# question04

population_million = 220

# Convert the 'Date' column to datetime format if not already
data['Date'] = pd.to_datetime(data['Date'])

# Calculate Per Thousand and Per Million stats
data['Tests per Thousand'] = (data['Daily Tests'] / population_million) / 1000
data['Cases per Thousand'] = (data['Daily Cases'] / population_million) / 1000
data['Recoveries per Thousand'] = (data['Daily Recoveries'] / population_million) / 1000
data['Deaths per Thousand'] = (data['Daily Deaths'] / population_million) / 1000

data['Tests per Million'] = (data['Daily Tests'] / population_million) * 1
data['Cases per Million'] = (data['Daily Cases'] / population_million) * 1
data['Recoveries per Million'] = (data['Daily Recoveries'] / population_million) * 1
data['Deaths per Million'] = (data['Daily Deaths'] / population_million) * 1

# Group by Month and aggregate the data
monthly_data = data.groupby(data['Date'].dt.to_period('M')).agg({
    'Daily Tests': 'sum',
    'Daily Cases': 'sum',
    'Tests per Thousand': 'sum',
    'Cases per Thousand': 'sum',
    'Tests per Million': 'sum',
    'Cases per Million': 'sum',
}).reset_index()

# Rename the 'Date' column to 'Month' and format it
monthly_data['Month'] = monthly_data['Date'].dt.strftime('%B')
monthly_data = monthly_data.drop(columns=['Date'])

# Calculate the positive rate
monthly_data['Positive Rate'] = (monthly_data['Daily Cases'] / monthly_data['Daily Tests']) * 100

# Rearrange the columns to match the format in the image
monthly_data = monthly_data[['Month', 'Daily Tests', 'Daily Cases', 'Tests per Thousand', 'Cases per Thousand', 'Tests per Million', 'Cases per Million', 'Positive Rate']]



# Display the DataFrame as a formatted table
styled_table = monthly_data.style.set_table_styles(
    [{'selector': 'thead th', 'props': [('background-color', '#2d4154'), ('color', 'white')]}]
).background_gradient(cmap='Blues', subset=['Daily Tests', 'Daily Cases', 'Tests per Thousand', 'Cases per Thousand', 'Tests per Million', 'Cases per Million', 'Positive Rate'])

# Show the styled DataFrame
styled_table
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Prepare data for ML models
X = data[['Daily Tests']]  # Feature (you can choose other features as well)
y_cases = data['Daily Cases']  # Target for cases
y_deaths = data['Daily Deaths']  # Target for deaths

# Split the data into training and testing sets
X_train, X_test, y_cases_train, y_cases_test = train_test_split(X, y_cases, test_size=0.2, random_state=42)
_, _, y_deaths_train, y_deaths_test = train_test_split(X, y_deaths, test_size=0.2, random_state=42)

# Linear Regression Model for Daily Cases
lr_model = LinearRegression()
lr_model.fit(X_train, y_cases_train)
cases_predictions = lr_model.predict(X_test)

# Random Forest Model for Daily Deaths
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_deaths_train)
deaths_predictions = rf_model.predict(X_test)

# Display the results
print(f"Linear Regression Model for Cases: {cases_predictions}")
print(f"Random Forest Model for Deaths: {deaths_predictions}")
```
